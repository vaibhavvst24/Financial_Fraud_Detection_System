# After the Label Encoding the values are 

```python
from sklearn.preprocessing import LabelEncoder

# Store mappings
label_mappings = {}

cat_cols = [
    'Merchant_Category',
    'Payment_Method',
    'Device_Type',
    'Location',
    'Suspicious_Keyword'
]

# Encoding + Saving Mapping
for col in cat_cols:

    le = LabelEncoder()

    df[col] = le.fit_transform(df[col])

    # Save mapping
    mapping = dict(zip(le.classes_, le.transform(le.classes_)))

    label_mappings[col] = mapping


# ==========================================
# PRINT ALL MAPPINGS
# ==========================================
for col, mapping in label_mappings.items():

    print(f"\n===== {col} =====")

    for category, label in mapping.items():
        print(f"{category} --> {label}")
```

# Fraud Percentage

```python
fraud_percentage = (
    df['Fraudulent'].mean()
) * 100

print(f"Fraud Percentage : {fraud_percentage}")
```

# Average Fraud Transaction Amount

```python
fta = df[df['Fraudulent'] == 1]['Transaction_Amount'].mean()
print(f"Average Fraud Transaction Amount : {fta:.3f}")
```

```python
df[df['Fraudulent'] == 0]['Transaction_Amount'].mean()
```

# Highest Fraudulent Transaction

```python
hft = df[df['Fraudulent'] == 1]['Transaction_Amount'].max()
print(f"High Fraud Transaction : {hft:.3f}")
```

# Payment Method Most Used in Fraud

```python
pm = df[df['Fraudulent'] == 1][
    'Merchant_Category'
].value_counts()

pmf = pd.DataFrame(pm)
pmf
```

# International Fraud Transactions

```python
ifs = df[
    (df['Is_International'] == 1) &
    (df['Fraudulent'] == 1)
].shape[0]

print(f"Number of International Fraud Transactions are : {ifs}")
```

# Fraud Rate in International Transactions

```python
international = df[
    df['Is_International'] == 1
]

rate = (
    international['Fraudulent'].mean()
) * 100

print(f"Fraud Rate in International Transactions are : {rate:.1f} %")
```

# Top Locations with Fraud

```python
locations = df[df['Fraudulent'] == 1][
    'Location'
].value_counts().head(10)

TL = pd.DataFrame(locations)
TL
```

# Customers with Highest Fraud Amount

```python
CH = df[df['Fraudulent'] == 1].groupby(
    'Customer_ID'
)['Transaction_Amount'].sum().sort_values(
    ascending=False
).head(10)

CHA = pd.DataFrame(CH)
print("Suspicious Customer")
CHA
```

# Fraud During Late Night

```python
df['Hour'] = pd.to_datetime(
    df['Transaction_Date'], format='%d-%m-%Y %H:%M',
    errors='coerce'
).dt.hour
```

```python
late_night_fraud = df[
    (df['Hour'] >= 0) &
    (df['Hour'] <= 5) &
    (df['Fraudulent'] == 1)
]

print(f"Fraud During Late Night : {late_night_fraud.shape[0]}")
```

# Transactions Above Average Spend

```python
ta = df[
    df['Transaction_Amount'] >
    df['Average_Spend']
].shape[0]

print(F"Transactions Above Average Spend are : {ta}")
```

### Fraud among them:

```python
fa = df[
    (df['Transaction_Amount'] >
     df['Average_Spend']) &
    (df['Fraudulent'] == 1)
].shape[0]

print(f"Fraud Among Them are : {fa}")
```

# Top Suspicious Keywords

```python
ts = df[df['Fraudulent'] == 1][
    'Suspicious_Keyword'
].value_counts()

tsk = pd.DataFrame(ts)
tsk
```

# Fraud Trend by Day

```python
df['Day'] = pd.to_datetime(
    df['Transaction_Date'], format='%d-%m-%Y %H:%M',
    errors='coerce'
).dt.day_name()

ft = df.groupby('Day')[
    'Fraudulent'
].sum().sort_values(
    ascending=False
)

ftd = pd.DataFrame(ft)
ftd
```

```python
df['Month'] = pd.to_datetime(
    df['Transaction_Date'], format='%d-%m-%Y %H:%M',
    errors='coerce'
).dt.month

fm = df.groupby('Month')[
    'Fraudulent'
].sum()

ftm = pd.DataFrame(fm)
ftm
```

# Correlation with Fraud

```python
numeric_df = df.select_dtypes(include='number')

cf = numeric_df.corr()['Fraudulent'].sort_values(
    ascending=False
)

cwf = pd.DataFrame(cf)
cwf
```
